```python
import numpy as np

# Parameters
n_samples = 100000
p0 = 0.1
p1 = 0.9

# Simulate Y
Y = np.random.choice([0, 1], size=n_samples, p=[p0, p1])

# Estimators
mean_estimator = np.mean(Y)  # Should be close to 0.9
median_estimator = np.median(Y)  # Should be 1

# Compute empirical absolute risk
risk_mean = np.mean(np.abs(Y - mean_estimator))
risk_median = np.mean(np.abs(Y - median_estimator))

print(f"Empirical absolute risk for mean estimator: {risk_mean:.4f}")
print(f"Empirical absolute risk for median estimator: {risk_median:.4f}")
```

The results are verified by the simulation.
